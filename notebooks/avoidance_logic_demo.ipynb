# 🚧 Obstacle Avoidance Logic Simulation
Visualizes how UAV reroutes when simulated obstacles are encountered.

```python
import matplotlib.pyplot as plt

original = [(0,0), (1,1), (2,2)]
reroutes = [(0,0), (1.01,1.01), (2,2)]

ox, oy = zip(*original)
rx, ry = zip(*reroutes)

plt.plot(ox, oy, 'o-', label='Original Path')
plt.plot(rx, ry, 'x--', label='Adjusted Path')
plt.title("Obstacle Avoidance Path Adjustment")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.grid(True)
plt.show()
```
